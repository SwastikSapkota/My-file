## Gravitational Force
$F=G\frac{m_1 m_2}{r^2}$
## Gravitational Potential
$V=G\frac{m_1}{r}$
## Gravitational Intensity
$g=G\frac{m_1}{r^2}$

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Gravitation():
    def __init__(self,G,m1,m2,r):
        self.G=G
        self.m1=m1
        self.m2=m2
    def Force(self,r):
        F=self.G*self.m1*self.m2/r**2
        return F
    def Potential(self,r):
        V=self.G*self.m1/r
        return V
    def Intensity(self,r):
        g=self.G*self.m1/r**2
        return g
```

```python
G=6.67*10**-11
m1=300 #in kg
m2=20 #in kg
R=[] #empty list to collect distance r
F=[] #empty list to collect force
V=[] #empty list to collect potential
g=[] #empty list to collect intensity
for r in range(1,10000):
    G1=Gravitation(G,m1,m2,r)
    f=G1.Force(r)
    v=G1.Potential(r)
    i=G1.Intensity(r)
    R.append(r)
    F.append(f)
    V.append(v)
    g.append(i)
    
```

```python
plt.figure(figsize= [10,10])

plt.subplot(2,2,1)
plt.plot(R,F,label='Force')
plt.plot(R,V,label='Potential')
plt.plot(R,g,label='Intensity')
plt.legend()
plt.xlabel('r')
plt.ylabel('F,V,g')

plt.subplot(2,2,2)
plt.semilogy(R,F,label='Force')
plt.semilogy(R,V,label='Potential')
plt.semilogy(R,g,label='Intensity')
plt.legend()
plt.xlabel('r')
plt.ylabel('ln F,ln V,ln g')

plt.subplot(2,2,3)
plt.loglog(R,F,label='Force')
plt.loglog(R,V,label='Potential')
plt.loglog(R,g,label='Intensity')
plt.legend()
plt.xlabel('ln r')
plt.ylabel('ln F,ln V,ln g')
```

```python
#Writing a suitable code for Electrostatic Force between two charges and Gravitational Force between two masses
def Fe(K,q1,q2,r):
    F1= (K*q1*q2)/r**2
    return F1

def Fg(G,m1,m2,r):
    F2=(G*m1*m2)/r**2
    return F2
```

```python
#calculating Fe and Fg between electron and proton with H-atom and comparing them
fE=Fe(9e9,1.6e-19,1.6e-19,5.29e-11)
fG=Fg(6.67e-11,9.1e-31,1.67e-27,5.29e-11)

if fE>fG:
    print("Electrostatic Force is greater than Gravitational Force")
else:
    print("Gravitational Force is greater than Electrostatic Force")
```
